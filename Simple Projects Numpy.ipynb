 Student Grades Analyzer
 This project analyzes student grades and calculates basic statistics.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample student grades
grades = [78, 85, 92, 67, 88, 76, 95, 80, 82, 91]
```

```python
mean = np.mean(grades)
median = np.median(grades)
mode = max(set(grades), key=grades.count)
std_dev = np.std(grades)
```

```python
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")
```

```python
plt.hist(grades, bins=5, edgecolor='black')
plt.title("Grade Distribution")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean}')
plt.axvline(mode, color='green', linestyle='dashed', linewidth=3, label=f'Mode: {mode}')
plt.axvline(median, color='orange', linestyle='dashed', linewidth=3, label=f'Median:{median}')
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

2. Simple Linear Regression Model
 This project implements a simple linear regression model to predict house prices based on size.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data (house size in sqft and price in $1000)
X = np.array([1500, 1800, 2400, 3000, 3500]).reshape(-1, 1)  # Independent variable (size)
y = np.array([400, 500, 600, 650, 700])  # Dependent variable (price)
```

```python
X = np.array([1500, 1800, 2400, 3000, 3500]).reshape(-1, 1)  # Independent variable (size)
y = np.array([400, 500, 600, 650, 700])  # Dependent variable (price)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
predictions = model.predict(X)
```

```python
plt.scatter(X, y, color='blue')  # Scatter plot of data
plt.plot(X, predictions, color='red')  # Regression line
plt.title("House Size vs Price")
plt.xlabel("Size (sqft)")
plt.ylabel("Price ($1000)")
plt.show()

# Print the regression coefficients
print(f"Coefficient: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```
